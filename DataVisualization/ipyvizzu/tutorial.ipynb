```python
import pandas as pd
from ipyvizzu import Chart, Data, Config, Style
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/avrabyt/Streamlit-ipyvizzu/main/Notebooks/FIFA%20World%20Cup%20Attendance.csv',dtype={"Year": str})
df = df[:-1]
df
```

```python
data = Data()
data.add_data_frame(df)
# type(data)

display(df)
```

### Style Sheet

```python
example_style = {
    "plot": {
            "yAxis": {
                "label": {
                    "fontSize": "1em",
                    "paddingRight": "1.2em",
                },
                "title": {"color": "#ffffff00"},
            },
            "xAxis": {
                "label": {
                    "angle": "2.5",
                    "fontSize": "1.1em",
                    "paddingRight": "0em",
                    "paddingTop": "1em",
                },
                "title": {"fontSize": "0.8em", "paddingTop": "2.5em"},
            },
        },
        "logo": {"width": "5em"},
    }
```

### Create a Chart Object

```python
chart = Chart(width="800px", height="500px")
chart.animate(data)

chart.animate(
    Config(
        {
            "x": "Year",
            "y": ["Total_Attendance","Hosts"],
            "label": "Matches",
            "title": "1. Area chart",
            "geometry": "area",
            "color": "Hosts",
            
        }
    ),
    Style(example_style
         ),
)
# print(type(chart))
snapshot1 = chart.store()
```

```python
chart.animate(
Config(
            {
                "split": True,
                
                "title": "2.Split Area",
            }
        )
)
snapshot2 = chart.store()
```

```python
chart.animate(
Config({"split": False,
        "title": "3.Bar chart",
        "geometry": "rectangle",
       
       })
)

snapshot3 = chart.store()

chart.animate(Data.filter("record.Hosts == 'Uruguay' || record.Hosts == 'Qatar' "),
         Config(
            {
                "title": "4. Zoom to Specific Elements",
            }
         ))

snapshot4 = chart.store()
```

```python
chart.animate(Data.filter(None),
    Config( {
            "x": ["Year","Matches"],
            "y": ["Total_Attendance"],
            "color": "Hosts",
            "size": ["Total_Attendance"],
            "label": "Hosts",
            "title": "Bubble Plot",
            "geometry": "circle",
        
        }),
              
        style={"duration": 2, "delay": 4},
)

snapshot5 = chart.store()
               
chart.animate(
    Config.bubble({
                    "size": ["Total_Attendance"],
                    "color": "Hosts",
                    "title": "5. Show Share of Components (%)",
            }))


snapshot6 = chart.store()
chart.animate( 
    Config.treemap(
            {
                "size": "Total_Attendance",
                "color": "Hosts",
                "title": "8. Treemap",
            }
        )

)
snapshot7 = chart.store()
```

```python
chart.animate(snapshot1)
chart.animate(snapshot2)
chart.animate(snapshot3)
chart.animate(snapshot4)
chart.animate(snapshot5)
chart.animate(snapshot6)
chart.animate(snapshot7)
```
