```python
data = []

for i in range(len(te)):
    
    price = te[i].거래금액.string.strip()
    built_yr = te[i].건축년도.string.strip()
    
    trade_yr = te[i].년.string.strip()
    trade_mon = te[i].월.string.strip()
    trade_day = te[i].일.string.strip()
                
    dong_name = te[i].법정동.string.strip()
    apt_name = te[i].아파트.string.strip()
    size = te[i].전용면적.string.strip()
    gu_code = te[i].지역코드.string.strip()

    total = [trade_yr, trade_mon, trade_day, price, built_yr, dong_name, apt_name, size, gu_code]
    data.append(total)
   
    
```
