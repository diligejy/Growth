# Histogram

```python
import matplotlib.pyplot as plt
from matplotlib import cm

%matplotlib inline
from matplotlib import rcParams
from matplotlib import font_manager, rc
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import numpy as np

###한글 폰트###
rcParams['axes.unicode_minus'] = False
#plt.rcParams['font.family'] = 'Malgun Gothic'

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('house_price1.csv', index_col=0)
```

```python
plt.hist(df['price_pyung'], bins=50, color='g') #bins 구간갯수
plt.ylabel('만원')
plt.grid()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2,tight_layout=True)
fig.set_size_inches(9, 3)
x = df['price_pyung']

# density=True 값을 통하여 Y축에 density를 표기할 수 있습니다.
bins=30
axs[0].hist(x, bins=bins, density=True, color='red', cumulative=True)
axs[1].hist(x, bins=bins, density=True, color='cyan')
axs[0].set_xlabel('만원')
axs[1].set_xlabel('만원')
axs[0].grid()
axs[1].grid()
plt.show()
```

```python
df1 = df.loc[df['dong_name'].isin(["서초동"])]
df2 = df.loc[df['dong_name'].isin(["반포동"])]

plt.hist(df1['price_pyung'], bins=50, color='g', alpha=0.5, label='서초동') 
plt.hist(df2['price_pyung'], bins=50, color='b', alpha=0.5, label='반포동') 
plt.xlabel('만원')
plt.legend()
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2,tight_layout=True)
fig.set_size_inches(9, 3)

x0 = df1['price_pyung']
x1 = df2['price_pyung']

bins=30
sns.histplot(x0, bins=bins, color='g',alpha=0.50, label='서초동', ax = axes[0])
sns.histplot(x1, bins=bins, color='b',alpha=0.50, label='반포동', ax = axes[1])
axes[0].grid()
axes[1].grid()
axes[0].legend()
axes[1].legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2,tight_layout=True)
fig.set_size_inches(9, 3)

x0 = df1['price_pyung']
x1 = df2['price_pyung']

bins=30
sns.kdeplot(x0, color='g',label='서초동', ax = axes[0])
sns.kdeplot(x1, color='b', label='반포동', ax = axes[1])
axes[0].grid()
axes[1].grid()
axes[0].legend()
axes[1].legend()
plt.savefig("house_price5.png")
plt.show()
```
