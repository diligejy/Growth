```python
import matplotlib.pyplot as plt
from matplotlib import cm

%matplotlib inline
from matplotlib import rcParams
from matplotlib import font_manager, rc
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import numpy as np

###한글 폰트###
rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('house_price1.csv')
```

```python
df.tail()
```

### matplotlib

```python
df.boxplot(column='price_pyung', by='dong_name', notch=True,  color='red', figsize=(8,5))
plt.xlabel("")
plt.title("")
plt.savefig("boxPlot1.png")
plt.show()
```

### seaborn

```python
sns.color_palette("rainbow", 8) # rainbow 색상표에서 8개의 색상을 가져옴
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x='dong_name', y='price_pyung', notch=True, palette="rainbow", data=df)
plt.xlabel('')
plt.ylabel('')
plt.grid()
plt.savefig("boxPlot2.png")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x='dong_name', y='price_pyung', notch=True, palette="rainbow", data=df)
sns.stripplot(x='dong_name', y='price_pyung', data=df, jitter=True, marker='o', alpha=0.3, color='red')
plt.xlabel('')
plt.ylabel('')
plt.grid()
plt.savefig("boxPlot3.png")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2,tight_layout=True)
fig.set_size_inches(9, 3)
sns.boxplot(x='dong_name', y='price_pyung', notch=True, palette="rainbow", data=df, ax = axes[0])

sns.boxplot(x='dong_name', y='price_pyung', notch=True, palette="rainbow", data=df, ax = axes[1])
sns.stripplot(x='dong_name', y='price_pyung', data=df, jitter=True, marker='o', alpha=0.3, color='red', s=4, ax = axes[1])

axes[0].set_xlabel('')
axes[0].set_ylabel('')
axes[0].grid()

axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].grid()
plt.savefig("boxPlot4.png")
plt.show()
```
