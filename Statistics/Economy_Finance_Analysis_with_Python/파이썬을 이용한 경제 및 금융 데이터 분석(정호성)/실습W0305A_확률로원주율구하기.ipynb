원의 면적을 알면 S= $\pi \times r^2$에 의해 원주율(π, pi, 파이)을 알 수 있다. 

무작위하게 수를 발생시켜 그것을 좌표로 점을 찍고, 점이 원의 영역에 포함되었는지의 여부를 가린다. 

이것을 반복하여 원에 포함된 점과 그렇지 않은 점의 수를 세고, 두 수의 비율을 통해 원의 면적을 구한다.

가로 1, 세로 1

직사각형 면적 : 1

반지름이 1인 원의 면적 $ \pi = \pi \times 1^2$  

반지름이 1인 원의 1/4의 원의 면적 $ \pi / 4$

원점에서 점까지의 거리 : 피타고라스 정리, $ \sqrt{x^2 + y^2}$

만약 임의의 점의 거리가 1보다 작으면 원 이내에 있음

따라서  $ \pi / 4 $ = (점이 1/4원에 있을 경우 수)/  (모든 점 수)  

 $ \pi $ = {(점이 1/4원에 있을 경우 수)/  (모든 점 수)} $\times$ 4
