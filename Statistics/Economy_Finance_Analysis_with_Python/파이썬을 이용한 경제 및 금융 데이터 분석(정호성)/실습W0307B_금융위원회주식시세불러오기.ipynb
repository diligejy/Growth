```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup

ServiceKey = "PQHpCPO8XCUXCPOF7nf1CW1ue7o92cGPp5Wurd%2F8Dc370C7cl2U3K%2F9MiRRWvDANczMiUiDNHyIPA%2F92YqiK9A%3D%3D"
stockCode = '005930'
numOfRows=10000
beginDay=20220101
endDay=20220625

url = "https://api.odcloud.kr/api/GetStockSecuritiesInfoService/v1/getStockPriceInfo?numOfRows="+str(numOfRows)+"&beginBasDt="+str(beginDay)+"&endBasDt="+str(endDay)+"&likeSrtnCd="+str(stockCode)+"&serviceKey="+str(ServiceKey)

result = urlopen(url) #this is for response of XML
stock = BeautifulSoup(result, 'lxml-xml')
te=stock.find_all("item")
```

```python
te
```

```python
data = []

for i in range(len(te)):
    
    date = te[i].basDt.string.strip()   
    market = te[i].mrktCtg.string.strip()       
    
    close = te[i].clpr.string.strip()
    open = te[i].mkp.string.strip()
    high = te[i].hipr.string.strip()
    low = te[i].lopr.string.strip()
    
    stockCode = te[i].srtnCd.string.strip()
    stockName = te[i].itmsNm.string.strip()
 
    volume = te[i].trqu.string.strip() # 거래량
    amount = te[i].trPrc.string.strip() #거래대금
    totShare = te[i].lstgStCnt.string.strip() #상장주식수
    totAmount = te[i].mrktTotAmt.string #시가총액    

    total = [date, stockCode, stockName, market, open, close, high, low, volume, amount, totShare, totAmount]
    
    data.append(total)  
    
```

```python
data
```

```python
temp = pd.DataFrame(data, columns= ["date", "stockCode", "stockName", "market", "open", "close", "high", "low", "volume", "amount", "totShare", "totAmount"])
```
