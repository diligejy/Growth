```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("household_2020a.csv")
```

```python
income = df['income'].dropna()
```

```python
income.describe()
```

# np.mean(income)

```python
np.round(np.mean(income),2)
```

```python
np.round(np.std(income),2)
```

```python
se = np.std(income) / np.sqrt(len(income))
```

```python
np.round(se,2)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.hist(income, alpha=0.3, bins=2000) 
ax.grid(True)
ax.set_xlim((-5000, 20000))
plt.savefig("assetAll_hist.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
n = 10000

mn = [] 

for i in range(n):
    sample = np.random.choice(income, size=500, replace=True)
    mean = np.mean(sample)
    mn.append(mean)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.hist(mn, alpha=0.3, bins=1000) 
ax.grid(True)
ax.set_xlim((3500, 7500))
plt.show()
```
