```python
import datetime
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

import xml.etree.ElementTree as ET 
```

```python
key='H4MNGW6FSSUM2SSYFHJO'
```

```python
url="http://ecos.bok.or.kr/api/StatisticTableList/"+str(key)+"/xml/kr/1/1000/" #1000개
```

```python
raw = requests.get(url)
xml = BeautifulSoup(raw.text,'xml')
```

```python
xml
```

```python
raw_data = xml.find_all("row")
```

```python
raw_data
```

```python
#raw_data[1]
```

```python
#raw_data[1].find('CYCLE').text
```

```python
data = []

for i in range(len(raw_data)):
    P_STAT_CODE = raw_data[i].P_STAT_CODE.string.strip()
    STAT_CODE = raw_data[i].STAT_CODE.string.strip()
    STAT_NAME = raw_data[i].STAT_NAME.string.strip()
    CYCLE = raw_data[1].find('CYCLE').text
    USED = raw_data[i].SRCH_YN.string.strip()
    ORG_NAME = raw_data[i].ORG_NAME.string

    total = [P_STAT_CODE, STAT_CODE, STAT_NAME, CYCLE, USED, ORG_NAME]
    data.append(total) 
```

```python
#data
```

```python
temp = pd.DataFrame(data,columns=["P_STAT_CODE", "STAT_CODE","STAT_NAME", "CYCLE", "USED", "ORG_NAME"]) 
```

```python
temp.to_csv('BOK_Total_LIST.csv', encoding='CP949')
```
