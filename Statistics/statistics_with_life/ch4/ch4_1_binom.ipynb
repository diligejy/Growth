```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import binom
```

```python
n = 5
p = 0.5 
```

```python
# 0 ~ (n)   ()
x = np.arange(n+1)
mean, var = binom.stats(n, p)
```

```python
prob = binom.pmf(x, n, p)
print('mean = ', mean, 'var = ', var)
```

```python
plt.bar(x, prob)
plt.xlabel("X")
plt.ylabel("P(X)")
plt.title("Binomial Distribution(n=5, p=0.5)")
plt.show()
```

 5      X  , X 3  (5     3 , P(X<=3)   )

->      ,           (cdf, cumulative distribution function) .

P(X <= 3) = P(X = 0) + P(X = 1) + P(X = 2) + P(X = 3)

```python
n = 5
p = 0.5

print('P(X<=3)=', binom.cdf(3, n, p))
```

```python
n = 100
p = 0.5
percent_point = 0.95

print("P(X<=k) = 0.95, k=", binom.ppf(percent_point, n, p))
```
