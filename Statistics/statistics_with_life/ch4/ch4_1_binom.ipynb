```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import binom
```

```python
n = 5
p = 0.5 
```

```python
# 0 ~ (n)까지의 리스트 생성 (확률변수)
x = np.arange(n+1)
mean, var = binom.stats(n, p)
```

```python
prob = binom.pmf(x, n, p)
print('mean = ', mean, 'var = ', var)
```

```python
plt.bar(x, prob)
plt.xlabel("X")
plt.ylabel("P(X)")
plt.title("Binomial Distribution(n=5, p=0.5)")
plt.show()
```

```python
print(prob)
```

### 문제 
동전을 5번 던졌을 경우 앞면이 나타난 횟수를 X라 할 때, X가 3보다 작을 확률(5회 시행에서 앞면이 나타난 횟수가 3 이하, P(X<=3)에 대한 확률을 계산하시오)

-> 확률질량함수는 특정 확률변수에 대한 확률을 계산하지만, 특정 확률변수의 범위나 특정 확률변수보다 같거나 작을 확률을 계산할 때는 누적분포함수(cdf, cumulative distribution function)를 이용합니다.

P(X <= 3) = P(X = 0) + P(X = 1) + P(X = 2) + P(X = 3)

```python
n = 5
p = 0.5

print('P(X<=3)=', binom.cdf(3, n, p))
```

### 문제 - 이항확률분포에 대한 누적분포함수의 역함수

동전을 100번 던졌을 경우 앞면이 나타난 횟수를 X라 할 때, P(X <= k) = 0.95를 만족하는 확률 변수 k를 계산하시오.





```python
n = 100
p = 0.5
percent_point = 0.95

print("P(X<=k) = 0.95, k=", binom.ppf(percent_point, n, p))
```
