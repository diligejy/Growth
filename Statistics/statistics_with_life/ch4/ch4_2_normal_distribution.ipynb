### EX-1

평균이 50이고 표준편차가 5인 정규분포 그래프 작성

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm

mu = 50
scale = 5

x = np.arange(mu-30, mu + 30, 0.1)
y = norm.pdf(x, mu, scale)

plt.bar(x, y)
plt.xlabel("X")
plt.ylabel("f(X)")
plt.title("Normal Distribution(mu=50, sigma=5)")
plt.show()
```
